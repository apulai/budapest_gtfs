```python
import folium
```

```python
m = folium.Map(location=[47.498364, 19.040602], zoom_start=12)
feature_group = folium.FeatureGroup(name="Icon collection", control=False).add_to(m)
```

```python
radius = 5
folium.CircleMarker(
    location=[47.498364, 19.040602],
    radius=radius,
    color="red",
    stroke=False,
    fill=True,
    fill_opacity=0.6,
    opacity=1,
    popup="{} pixels".format(radius),
    tooltip="I am in pixels",
).add_to(feature_group)
folium.CircleMarker(
    location=[47.48657049786258, 18.953883647694855],
    tooltip="Click me!",
    popup="KFKI",
    icon=folium.Icon(icon="cloud"),
).add_to(feature_group)

folium.CircleMarker(
    location=[47.598877, 19.352266],
    tooltip="Click me!",
    popup="Szabadsagter",
    icon=folium.Icon(color="red"),
).add_to(feature_group)
m
```
